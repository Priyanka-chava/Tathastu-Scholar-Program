```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime 
import wikipedia 
import pyjokes 
import sys
import webbrowser
from datetime import date

```

```python
engine=pyttsx3.init() 
engine.setProperty("rate", 150) 
voices = engine.getProperty("voices")
engine.setProperty('voice', voices [1].id) 
recognizer=sr.Recognizer()

```

```python
def engine_talk(text):
    engine.say(text) 
    engine.runAndWait()

```
